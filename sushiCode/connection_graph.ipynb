```python
def cal_pair_in_common(vMap):
    com_mat = np.zeros((len(vMap),len(vMap)), dtype = np.int8)
    for i in vMap.keys():
        for j in vMap.keys():
            com_element=len(set(vMap[i].get_order_b()).intersection(set(vMap[j].get_order_b())))
            #com_mat[i][j] = com_element*(com_element-1)/2
            com_mat[i][j] = com_element
    return com_mat
common_pair_mat = cal_pair_in_common(vMap)
```

```python
g = nx.Graph()

n = common_pair_mat.shape[0]

g.add_nodes_from(list(range(n)))

for i in range(n):
    for j in range(n):
        if i<j:
            #g.add_node((i,j))
            if common_pair_mat[i][j]>4:
            #print(i,j)
                g.add_edge(i,j, weight = common_pair_mat[i][j])
nx.number_of_nodes(g)
```
